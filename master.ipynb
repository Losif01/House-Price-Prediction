```python
def getMissingValues(df):
    missing_values = df.isnull().sum().sort_values(ascending=False)
    missing_values = missing_values[missing_values > 0]
    missing_values = missing_values / len(df)
    return f" {missing_values} \n\n Number of missing values: {missing_values.__len__()}"
```

```python
print(getMissingValues(df_train))
print(getMissingValues(df_test))
```

```python
def getUniqueVaules(df):
    unique_values = []
    for i in df.columns:
        unique_values.append((i , df[i].nunique()))
    return f" {unique_values} "

getUniqueVaules(df_train[["BsmtCond", "BsmtQual", "BsmtExposure", "BsmtFinType1", "BsmtFinType2"]])
```
