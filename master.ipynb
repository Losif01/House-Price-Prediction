```python
def getMissingValues(df):
    missing_values = df.isnull().sum().sort_values(ascending=False)
    missing_values = missing_values[missing_values > 0]
    missing_values = missing_values / len(df)
    return [missing_values], missing_values.__len__()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plotBsmt(data):
    # Separate the dictionary keys and values
    keys = list(data.keys())
    values = list(data.values())

# Plot the bar histogram
    plt.figure(figsize=(10, 6))
    plt.bar(range(len(data)), values, tick_label=keys)

# Add title and labels
    plt.title('Bar Histogram of Dictionary Data')
    plt.xlabel('Keys')
    plt.ylabel('Values')

    # Display the plot
    plt.show()
```

```python
plotBsmt(Bsmtcond) # fill with TA
plotBsmt(Bsmtqual) # fill randomly with Gd and TA
plotBsmt(Bsmtxp)   # fill with No
plotBsmt(Bsmtt1)   # fill randomly with controlled probabilities
plotBsmt(Bsmtt2)   # fill with Unf
```
